```python
%pylab inline
from sklearn import datasets

X, y = datasets.make_moons(200, noise=.05, random_state=0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap="viridis");
```

### 1.1 Running K-means

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
kmeans.labels_
```

```python
import numpy as np
sum(np.logical_and(y_kmeans, y))/y.shape[0]
```

### 1.2 Visualizing results

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='blue', s=200, alpha=0.5)
```

#### Compare Results:

##Before using kmeans, the two sets of values are differiented obviously. After using kmeans, they are interacted 
with each other. And there are two center points for each set.

#### Use Different K Values:

```python
## k=5
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

### 2 Hierarchical Clustering

 #### Dataset Description:  Wine dataset in scikit-learn is used for this question, Alcohol Malic acid are used for analysis

```python
from sklearn.datasets import load_wine
wine = load_wine()
X = wine.data[:,:2]
y = wine.target
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap="viridis");
```

### 2.1 Running Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
res = AgglomerativeClustering().fit(X)
ret = res.fit_predict(X)
res.labels_
```

```python
sum(np.logical_and(ret, y))/y.shape[0]
```

### 2.2 Visualizing Results

```python
plt.scatter(X[:, 0], X[:, 1], c=ret, s=50, cmap='viridis')
```

#### Compare Results:

##Before using kmeans, there are three kinds of values and the points are scatterd. 
After using kmeans, there are only tow kinds of values and the points seem more sortable.

#### Use Different K Values:

```python
## k=3
res_k3 = AgglomerativeClustering(n_clusters=3).fit(X)
ret_k3 = res_k3.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=ret_k3, s=50, cmap='viridis')
```

```python
## k=4
res_k4 = AgglomerativeClustering(n_clusters=4).fit(X)
ret_k4 = res_k4.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=ret_k4, s=50, cmap='viridis')
```
