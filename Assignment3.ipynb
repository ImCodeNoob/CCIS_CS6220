```python
## 2.1 Frequent Itemset Generation
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

groceries_dataset = [[s.strip() for s in l.split(",")] 
                     for l in open("/Users/wufeiyue/Downloads/assignment3_datasets/groceries.csv").readlines()]

te = TransactionEncoder()
te_ary = te.fit(groceries_dataset).transform(groceries_dataset)
df1 = pd.DataFrame(te_ary, columns=te.columns_)
df1.head()
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
fre_item_df= apriori(df1,min_support = 0,use_colnames=True,max_len = 2)
fre_item_df['length'] = fre_item_df['itemsets'].apply(lambda x: len(x))
fre_item_df
fre_item_df1 = fre_item_df[(fre_item_df['length'] == 1)]
fre_item_df2 = fre_item_df[fre_item_df['length'] == 2]
```

```python
## Top 5 1-itemsets with the highest support
```

```python
fre_item_df1.sort_values(by=['support'],ascending=False).head()
```

```python
## The highest support value for the 1-itemsets
```

```python
fre_item_df1.sort_values(by=['support'],ascending=False).head(1)
```

```python
## The top 5 2-itemsets with the highest support
```

```python
fre_item_df2.sort_values(by=['support'],ascending=False).head()
```

```python
## The highest support value for the 2-itemsets
```

```python
fre_item_df2.sort_values(by=['support'],ascending=False).head(1)
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
fre_item_df2= apriori(df1,min_support =0.04,use_colnames=True,max_len = 2)
```
